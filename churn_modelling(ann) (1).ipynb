```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```

```python
df.describe
```

```python
x=df.iloc[:,3:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x['Geography'].replace({'France':0,'Germany':1,'Spain':2},inplace=True)
```

```python
x['Gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from  sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential  # it is used to initialize ann
from tensorflow.keras.layers import Dense # provides layer
```

```python
ann=Sequential() # step1 initialize the ann

ann.add(Dense(units=6,activation='relu'))  # step2 add hidden layer

ann.add(Dense(units=1,activation='sigmoid')) # step add output layer

ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #establishing connection

ann.fit(xtrain,ytrain,batch_size=30,epochs=50) #train the model

ypred=ann.predict(xtest)# make prediction
```

```python
ypred=ypred>0.5
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
